```python
for i in loaded_TX1:
    driver.get(i)
    time.sleep(1)
    try:
        add=driver.find_element(By.XPATH,'//ul[@class="breadcrumbs"]/li[4]/span')
        adress3.append(add.text)
    except NoSuchElementException:
        adress3.append('null')
    try:
        b=driver.find_element(By.XPATH,'//div[@class="listing-detail-stat-container w-1/2 md:w-1/4"]/h4')
        bad3.append(b.text)
    except NoSuchElementException:
        bad3.append('null')
    try:
        ba=driver.find_element(By.XPATH,'//div[@class="listing-detail-stat-container w-1/2 md:w-1/4"][2]/h4')
        bath3.append(ba.text)
    except NoSuchElementException:
        bath3.append('null')
    try:
        a=driver.find_element(By.XPATH,'//div[@class="listing-detail-stat-container w-1/2 md:w-1/4"][4]/h4')
        Area3.append(a.text)
    except NoSuchElementException:
        Area3.append('null')
    try:
        p=driver.find_element(By.XPATH,'//div[@class="main-data-section-header"][2]/div[1]/div[2]')
        parking3.append(p.text)
    except NoSuchElementException:
        parking3.append('null')
    try:
        ac=driver.find_element(By.XPATH,'//div[@class="main-data-section-header"][5]/div[1]/div[2]')
        AC3.append(ac.text)
    except NoSuchElementException:
        AC3.append('null')    
    try:
        ap=driver.find_element(By.XPATH,'/html/body/div[1]/div/div/main/article/div[4]/div[2]/div[1]/section[1]/section/div[2]/div/div/div[1]/div[2]/div[1]/div/div[2]')
        Appliances3.append(ap.text)
    except NoSuchElementException:
        Appliances3.append('null')    
    try:
        ct=driver.find_element(By.XPATH,'//ul[@class="breadcrumbs"]/li[2]/a')
        City3.append(ct.text)
    except NoSuchElementException:
        City3.append('null')  
    try:
        z=driver.find_element(By.XPATH,'//ul[@class="breadcrumbs"]/li[3]/a')
        zip_code3.append(z.text)
    except NoSuchElementException:
        zip_code3.append('null') 
    try:
        add=driver.find_element(By.XPATH,'//h4[@class="md:text-right h3"]')
        price2.append(add.text)
    except NoSuchElementException:
        price2.append('null')
```
